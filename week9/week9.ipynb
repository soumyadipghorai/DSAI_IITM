Grammer of graphics:

data -> aesthetics -> scale -> geometric objects -> statistics -> facets -> coordinate system

```python
import plotly.express as px
import pandas as pd

df = px.data.gapminder()
df.head()
```

```python
fig = px.scatter(
    data_frame=df[df['year'] == 2007],
    x = 'gdpPercap',
    y = 'lifeExp',
    color = 'continent',
    size = 'pop',
    hover_name='country',
    log_x = True,
    size_max = 60,
    color_discrete_sequence=px.colors.qualitative.Plotly,
    title = 'GDP per capita vs life expectancy'
)
fig.update_layout(template = 'plotly_dark')
fig.show()
```

```python
df_india = df[df['country'] == 'India']
fig = px.line(
    data_frame= df_india,
    x = 'year',
    y='gdpPercap',
    markers = True,
    color_discrete_sequence=['magenta'],
    title = 'GDP per capita over time - Inida',

)
fig.update_layout(
    template = 'plotly_dark',
    xaxis_title = 'year',
    yaxis_title = 'GDP per capita'
)
fig.show()
```

```python
df_bar = df[df['year'] == 2007].groupby('continent', as_index = False)['lifeExp'].mean()

fig = px.bar(
    data_frame = df_bar,
    x = 'continent', y = 'lifeExp', color = 'continent',
    color_discrete_sequence=px.colors.qualitative.Dark2,
    title = 'Avg life expectancy by continent'
)

fig.update_layout(
    template = 'plotly_dark', xaxis_title = 'continent', yaxis_title = ' avg life expectancy'
)
fig.show()
```

```python
fig = px.box(
    df[df['year'] == 2007], x = 'continent', y = 'lifeExp',
    color = 'continent', color_discrete_sequence=px.colors.qualitative.Dark2,
    title = 'Box plot - life expectancy by continent'
)
fig.update_layout(template = 'plotly_dark')
fig.show()
```

```python
fig = px.violin(
    df[df['year'] == 2007], x = 'continent', y = 'lifeExp',
    color = 'continent', color_discrete_sequence=px.colors.qualitative.Dark24,
    title = 'violin plot - life expectancy by continent',
    box = True, points = 'all'
)
fig.update_layout(template = 'plotly_dark')
fig.show()
```

```python
fig = px.scatter_matrix(
    df[df['year'] == 2007], dimensions=['lifeExp', 'gdpPercap', 'pop'],
    color = 'continent', color_discrete_sequence=px.colors.qualitative.Plotly,
    title = 'scatter matrix - life expectanc, GDP and population'
)
fig.update_traces(diagonal_visible = False)
fig.update_layout(template = 'plotly_dark')
fig.show()
```

```python
gdp_corr = df[['lifeExp', 'gdpPercap', 'pop']].corr()
fig = px.imshow(
    gdp_corr, text_auto = True, color_continuous_scale='Inferno',
    title = 'Correlation heatmap'
)
fig.update_layout(
    template = 'plotly_dark', xaxis_title = 'variable', yaxis_title = 'variable'
)
fig.show()
```

```python
fig = px.choropleth(
    df[df['year'] == 2007], locations = 'iso_alpha', color = 'lifeExp',
    hover_name = 'country', color_continuous_scale= 'Inferno',
    title = 'Global life expectancy'
)
fig.update_layout(template = 'plotly_dark')
fig.show()
```
